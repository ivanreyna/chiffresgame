```python
import random
# NUMBER GAME FUNCTIONS 
# target number of 3 digits
def target_number(a, b):
    return random.randint(a, b)
# Generate six random numbers
def random_numbers(a, b):
    digit1 = random.randint(a, b)
    digit2 = random.randint(a, b)
    digit3 = random.randint(a, b)
    digit4 = random.randint(a, b)
    digit5 = random.randint(a, b)
    digit6 = random.choice([25, 50, 75, 100])
    return digit1, digit2, digit3, digit4, digit5, digit6
# addition
def addition(x, y):
      return x + y
# subtraction
def subtraction(x, y):
    return x - y
# multiplication
def multiplication(x, y):
    return x * y
# division
def division(x, y):
    # Check division is not over zero
    if y == 0:
        return "Error: Division by 0"
    return x / y
# Perform the operation desided by the user
def evaluate(n1, operator, n2):
    if operator == '+':
        return addition(n1, n2)
    elif operator == '-':
        return subtraction(n1, n2)
    elif operator == '*':
        return multiplication(n1, n2)
    elif operator == '/':
        return division(n1, n2)
    else:
        return "Invalid operator"
# Ask if there are more operations to do
def continue_operation(result):
    while True:
        choice = input("More Operations ? (yes/no): ").lower()
        if choice == "yes":
            operator = input("Enter the next operator (+, -, *, /): ")
            num = int(input("Enter the next number: "))
            evaluated_result = evaluate(result, operator, num)
            print(f"New result: {evaluated_result}")
        elif choice == "no":
            return result
        else:
            print("Please enter 'yes' or 'no'.")
#Difference betwween user's estimation and number and the score 
def calculate_score(target, result):
    difference = abs(target - result)
    score = 100 - (difference // 10)
    return score

# WORD GAME

# Vowels and consonants
vowels = 'AEIOU'
consonants = 'BCDFGHJKLMNPQRSTVWXYZ'

# Look for the longest word from the dictionary based on available letters
def longest_word(letters, word_list):
    longest_word = ""
    letters = sorted(letters)

    # Check in the list if char in letters can form biggest word (from list)
    for w in word_list:
        w = w
        word_letters = sorted(w)
        
        # Check if the word can be formed from the provided letters
        if all(word_letters.count(char) <= letters.count(char) for char in word_letters):
            if len(w) > len(longest_word):
                longest_word = w

    # if longest_word just affirms there is a valid output from the list (to avoid bugs, zied just explained last class)
    return longest_word if longest_word else "No valid word found"

# Calculate score with S = 100 - (L - l) * 10
def calculate_word_score(player_word, longest_word):
    L = len(longest_word)
    l = len(player_word)
    return max(min(100 - (L - l) * 10, 100), 0)  # Limits scores from 0-100 (it was bugged, solved it in yt)

# Function to get the letters from the user
def generate_letters():
    letters = []
    for iteration in range(10): #will just diplay 10 times
        choice = input("Enter'v' or 'c' (vowel/consonant) ").strip().lower()
        if choice == 'v':
            letters.append(random.choice(vowels))
        elif choice == 'c':
            letters.append(random.choice(consonants))
        else:
            print("Please enter 'v' or 'c' (vowel/consonant).")
            continue  # WOnt advance if still wrong input
    return letters

# Load the list of words from a local file
def load_word_list(): #yt video as well
    file_path = "words.txt"  
    try:
        with open(file_path, 'r') as file:
            word_list = file.read().splitlines()  # Split the text into a list of words
        print(f"Loaded {len(word_list)} words from the local file.")
        return word_list
    except FileNotFoundError:
        print(f"File not found at {file_path}")
        return []  # Empty list if the file is not found
    
def play_number_game():
    print("The Number Game")
    target = target_number(100, 999)
    numbers = random_numbers(1, 10)
    print(f"Target number: {target}")
    print(f"Available numbers: {numbers}")
    result = int(input("Choose a number from the list belowv: "))
    result = continue_operation(result)
    score = calculate_score(target, result)
    print(f"Your result: {result}")
    print(f"Target was: {target}")
    print(f"Your score: {score}")

def play_word_game():
    print("The Word Game")
    letters = generate_letters()
    print(f"Your letters are: {''.join(letters)}")
    
    word_list = load_word_list()
    if not word_list:
        print("Unable to play the game due to missing word list.")
        return
    
    player_word = input("Enter your word: ").upper()
    
    longest_w = longest_word(letters, word_list)
    print(f"The longest possible word is: {longest_w}")
    
    score = calculate_word_score(player_word, longest_w)
    print(f"Your word: {player_word}")
    print(f"Your score: {score}")

# Run the games
if __name__ == "__main__":
    print("Number Game")
    play_number_game()

    print("\n Word Game")
    play_word_game()
```
